```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
trX = np.linspace(-1, 1, 101)
trY = 5 * trX + np.random.randn(*trX.shape) * 0.33 # y = 3x+ noise
```

```python
import matplotlib.pyplot as plt
plt.plot(trX,trY,'o')
plt.show()
```

```python
model = Sequential()
model.add(Dense(1, activation="linear"))
model.compile(optimizer='sgd', loss='mse')
#model.layers.append(Dense(1, activation='linear'))
model.fit(trX.reshape(-1,1), trY.reshape(-1,1),epochs=1)
model.summary()
```

```python
weights = model.layers[0].get_weights()
print((weights))
w_init = weights[0]
b_init = weights[1]

#y= 0.399 x - 0.004522
```

```python
plt.plot(trX,trY,'o')
plt.plot(trX,model.predict(trX),'r-')
plt.show()
```

```python

model.fit(trX.reshape(-1,1), trY.reshape(-1,1),epochs=200)
```

```python
weights = model.layers[0].get_weights()
print(np.size(weights))
w_final = weights[0]
b_final = weights[1]
```

```python
print('Linear regression model is initialized with weights w: %.2f, b: %.2f' % (w_final, b_final))
#y = 4.98 x +0.04
```

```python
plt.plot(trX,trY,'o')
plt.plot(trX,model.predict(trX))
plt.plot(trX,(trX*w_init+b_init).T,'r')

plt.show()
```

```python
model.save_weights('my_model.h5')
```
