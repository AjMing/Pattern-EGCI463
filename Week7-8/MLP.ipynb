```python
import numpy
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
X=numpy.array([ [0, 0] ,[0, 1], [1, 0],[1, 1]])
y=numpy.array([ 0, 1,1,0])
```

```python
mlp = MLPClassifier(verbose=1)
mlp.fit(X, y)
```

```python
mlp.predict(X)
```

```python
mlp.score(X,y)
```

```python
mlp.coefs_
```

```python
len(mlp.coefs_[1])
```

```python
mlp.coefs_[0]
```

```python

mlp = MLPClassifier(hidden_layer_sizes=(3,) ,activation='logistic',learning_rate_init=0.2,solver='sgd',max_iter=10000,tol=0.000001,verbose=1,random_state=10)
mlp.fit(X,y)
```

```python
mlp.predict_proba(X)
```

```python
mlp.predict(X)
```

```python
mlp.score(X,y)
```

```python
mlp.coefs_
```

```python
mlp.coefs_[0]
```

```python
mlp.coefs_[1]
```

```python
mlp.classes_
```

```python
mlp.intercepts_
```

```python
mlp.loss_curve_
```

```python
y=numpy.array([ 0, 1,1,0])
mlp = MLPClassifier(activation='tanh',max_iter=10000,hidden_layer_sizes=(3,),solver='sgd',validation_fraction=0)
mlp.fit(X, y)
mlp.predict(X)
```

```python
import matplotlib.pyplot as pt
pt.plot(mlp.loss_curve_)
pt.show()
```

```python
print(len(mlp.loss_curve_))
```

```python
mlp2.predict(X)
```
